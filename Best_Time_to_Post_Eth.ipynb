```python
## lets convert date to datetime format
df['date'] = pd.to_datetime(df['date'])
```

```python
## Defining a function to categorize the times of the day
## This simplifies our analysis and helps in identifying patterns

def categorize_time(hour):
    if 0 <= hour < 3:
        return '0-3'
    elif 3 <= hour < 6:
        return '3-6'
    elif 6 <= hour < 9:
        return '6-9'
    elif 9 <= hour < 12:
        return '9-12'
    elif 12 <= hour < 15:
        return '12-15'
    elif 15 <= hour < 18:
        return '15-18'
    elif 18 <= hour < 21:
        return '18-21'
    else:
        return '21-24'
```

```python
### Features Engineering : extracting new columns from existing ones
df['day'] = df['date'].dt.day_name() ## Extract day from date
df['time_range'] = df['date'].dt.hour.apply(categorize_time) ## Extract time range
df['has_image'] = df['format'].apply(lambda x: 'yes' if isinstance(x, str) and 'image' in x.lower() else 'no')
df['has_comarketing'] = df['type'].apply(lambda x: 'yes' if isinstance(x, str) and 'Co-marketing' in x else 'no')
df['engagement_rate'] = (df['retweets'] + df['replies'] + df['likes'] + df['detail_expands'] + df['follows']+df['media_views'] + df['media_engagements'] ) /df['impressions']
df['CTR'] = (df['profile_clicks'] + df ['url_clicks'] + df['hashtag_clicks']) / df['impressions']
```

### Nethermind_Eth

```python
## Extracting posts for NethermindEth alone

nethermind_eth = df[df['twitter_handle']== 'NethermindEth']
nethermind_eth.head()
```

```python
### Let's check the number of posts for Eth account per day
nethermind_eth.day.value_counts()
```

```python
### Saturday being an extreme outlier will be excluded from further analysis
### And weights will be applied to other days to eliminate bias

nethermind_eth = nethermind_eth[nethermind_eth['day'] != 'Saturday']
```

```python
# Calculate the number of posts per group
post_counts = nethermind_eth.groupby(['day']).size().reset_index(name='counts')

# Merge weights back to the original dataframe
nethermind_eth = nethermind_eth.merge(post_counts, on=['day'])
nethermind_eth['weight'] = nethermind_eth['counts'].apply(lambda x: 1 / x)
nethermind_eth['normalized_weight'] = nethermind_eth['weight'] / nethermind_eth['weight'].sum()

# Apply weights to engagement rate,Impressions and CTR
nethermind_eth['weighted_engagement_rate'] = nethermind_eth['engagement_rate'] * nethermind_eth['normalized_weight']
nethermind_eth['weighted_CTR'] = nethermind_eth['CTR'] * nethermind_eth['normalized_weight']
nethermind_eth['weighted_impressions'] = nethermind_eth['impressions'] * nethermind_eth['normalized_weight']


# Extract relevant fields
nethermind_eth= nethermind_eth[['day', 'time_range', 'has_image','has_comarketing', 'engagement_rate', 'CTR','impressions', 'weighted_engagement_rate', 'weighted_CTR','weighted_impressions']]
nethermind_eth.head()
```

```python
## Let's check the nature of our data (Test of normality & Variance)
## This will help us determine what type of statistical test to run and what assumptions will be made
## We will be checking for Normality & Variance

#Check for Normality: does our data group assume a normal distribution
```

```python
#Check for Normality
from scipy.stats import shapiro
import pandas as pd
grouped_data = nethermind_eth.groupby('day')

# Perform Shapiro-Wilk test for each group
for group_name, group_data in grouped_data:
    print(f"Day: {group_name}")

    # Shapiro-Wilk test for 'engagement_rate'
    stat, p_engagement = shapiro(group_data['engagement_rate'])
    print(f"Shapiro-Wilk Test for Engagement Rate - Statistics: {stat}, p-value: {p_engagement}")

    # Shapiro-Wilk test for 'CTR'
    stat, p_ctr = shapiro(group_data['CTR'])
    print(f"Shapiro-Wilk Test for CTR - Statistics: {stat}, p-value: {p_ctr}")

    # Shapiro-Wilk test for 'impressions'
    stat, p_impr = shapiro(group_data['impressions'])
    print(f"Shapiro-Wilk Test for CTR - Statistics: {stat}, p-value: {p_ctr}")

    # Check p-values for normality
    alpha = 0.05
    if p_engagement > alpha and p_ctr > alpha and p_impr > alpha:
        print("All metrics are normally distributed.")
    else:
        print("Normality does not apply.")
    print("\n")
```

```python
from scipy.stats import levene
grouped_data = nethermind_eth.groupby('day')

# Prepare lists to hold data from each group
engagement_data = []
ctr_data = []
impr_data = []

for group_name, group_data in grouped_data:
    engagement_data.append(group_data['engagement_rate'])
    ctr_data.append(group_data['CTR'])
    impr_data.append(group_data['impressions'])


# Perform Levene's test across all groups
stat_engagement, p_engagement = levene(*engagement_data)
stat_ctr, p_ctr = levene(*ctr_data)
stat_impr,  p_impr = levene(*impr_data)

print(f"Levene's Test for Engagement Rate - Statistics: {stat_engagement}, p-value: {p_engagement}")
print(f"Levene's Test for CTR - Statistics: {stat_ctr}, p-value: {p_ctr}")
print(f"Levene's Test for CTR - Statistics: {stat_impr}, p-value: {p_impr}")


# Check p-values for homogeneity of variances
alpha = 0.05
if p_engagement > alpha and p_ctr > alpha and p_impr > alpha:
    print("All metrics have homogeneous variances.")
else:
    print("homogeneous variances do not apply")
```

```python
#Check for Significance amongst all groups

from scipy.stats import kruskal

# Prepare data for Kruskal-Wallis H test
groups_engagement_rate = [nethermind_eth[nethermind_eth['day'] == day]['weighted_engagement_rate'] for day in nethermind_eth['day'].unique()]
groups_CTR = [nethermind_eth[nethermind_eth['day'] == day]['weighted_CTR'] for day in nethermind_eth['day'].unique()]
groups_impr = [nethermind_eth[nethermind_eth['day'] == day]['weighted_impressions'] for day in nethermind_eth['day'].unique()]


# Perform Kruskal-Wallis H test for engagement rate
statistic_er, p_value_er = kruskal(*groups_engagement_rate)
print("Kruskal-Wallis H statistic (Engagement Rate):", statistic_er)
print("p-value (Engagement Rate):", p_value_er)

# Perform Kruskal-Wallis H test for CTR
statistic_ctr, p_value_ctr = kruskal(*groups_CTR)
print("Kruskal-Wallis H statistic (CTR):", statistic_ctr)
print("p-value (CTR):", p_value_ctr)

# Perform Kruskal-Wallis H test for Impressions
statistic_ctr, p_value_impr = kruskal(*groups_impr)
print("Kruskal-Wallis H statistic (CTR):", statistic_ctr)
print("p-value (Impressions):", p_value_impr)



# Interpret results
if p_value_er < 0.05:
    print("Reject the null hypothesis for Engagement Rate. There is a significant difference among the groups.")
else:
    print("Fail to reject the null hypothesis for Engagement Rate. There is no significant difference among the groups.")

if p_value_ctr < 0.05:
    print("Reject the null hypothesis for CTR. There is a significant difference among the groups.")
else:
    print("Fail to reject the null hypothesis for CTR. There is no significant difference among the groups.")


if p_value_impr < 0.05:
    print("Reject the null hypothesis for Impressions. There is a significant difference among the groups.")
else:
    print("Fail to reject the null hypothesis for Impressions. There is no significant difference among the groups.")
```

```python
#Check for Significance amongst specific group
from scikit_posthocs import posthoc_dunn
# Perform Dunn's test for engagement rate
dunn_er = posthoc_dunn(nethermind_eth, val_col='weighted_engagement_rate', group_col='day', p_adjust='bonferroni')
print(dunn_er)

# Perform Dunn's test for CTR
dunn_ctr = posthoc_dunn(nethermind_eth, val_col='weighted_CTR', group_col='day', p_adjust='bonferroni')
print(dunn_ctr)


# Perform Dunn's test for Impressions
dunn_impr = posthoc_dunn(nethermind_eth, val_col='weighted_impressions', group_col='day', p_adjust='bonferroni')
print(dunn_impr)
```

```python
### Looking at the Image ratio & co-market ratio to understand possible cause of increased metrics

def image_ratio(group):
    total_posts = len(group)
    posts_with_image = group['has_image'].str.lower().eq('yes').sum()
    return posts_with_image / total_posts

# Function to calculate the ratio of posts with Co-marketing
def comarket_ratio(group):
    total_posts = len(group)
    posts_with_marketing = group['has_comarketing'].str.lower().eq('yes').sum()
    return posts_with_marketing / total_posts

# Group by day and calculate the ratio of posts with images
image_ratio_per_day = nethermind_eth.groupby('day').apply(image_ratio).reset_index(name='image_ratio')

# Group by day and calculate the ratio of posts with Co-marketing
comarket_ratio_per_day = nethermind_eth.groupby('day').apply(comarket_ratio).reset_index(name='comarket_ratio')

# Merge the two DataFrames on 'day'
combined_ratios_per_day = pd.merge(image_ratio_per_day, comarket_ratio_per_day, on='day')
combined_ratios_per_day
```
